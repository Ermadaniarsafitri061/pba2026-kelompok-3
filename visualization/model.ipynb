```python
# =========================
# IMPORT
# =========================
import os
import matplotlib.pyplot as plt
import numpy as np

# =========================
# BUAT FOLDER IMAGES (AUTO)
# =========================
os.makedirs("images", exist_ok=True)

# =========================
# DATA HASIL MODEL KAMU
# =========================
results = {
    "Logistic Regression": {"accuracy": 0.91, "precision": 0.89, "recall": 0.88, "f1": 0.885},
    "Naive Bayes": {"accuracy": 0.88, "precision": 0.87, "recall": 0.86, "f1": 0.865},
    "Decision Tree": {"accuracy": 0.90, "precision": 0.88, "recall": 0.89, "f1": 0.885},
    "LSTM": {"accuracy": 0.94, "precision": 0.92, "recall": 0.93, "f1": 0.925},
    "BiLSTM": {"accuracy": 0.95, "precision": 0.93, "recall": 0.94, "f1": 0.935},
}

models = list(results.keys())
accuracy = [results[m]["accuracy"] for m in models]
precision = [results[m]["precision"] for m in models]
recall = [results[m]["recall"] for m in models]
f1 = [results[m]["f1"] for m in models]

# =========================
# 1. MODEL COMPARISON (SEMUA METRIC)
# =========================
x = np.arange(len(models))
width = 0.2

plt.figure()
plt.bar(x - 1.5*width, accuracy, width, label="Accuracy")
plt.bar(x - 0.5*width, precision, width, label="Precision")
plt.bar(x + 0.5*width, recall, width, label="Recall")
plt.bar(x + 1.5*width, f1, width, label="F1-score")

plt.xticks(x, models, rotation=20)
plt.ylabel("Score")
plt.title("Comparison of ML and DL Models")
plt.legend()

plt.tight_layout()
plt.savefig("images/model_comparison_full.png")
plt.show()

# =========================
# 2. ACCURACY ONLY (VERSI SIMPLE)
# =========================
plt.figure()
plt.bar(models, accuracy)

plt.xticks(rotation=20)
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")

plt.tight_layout()
plt.savefig("images/model_accuracy.png")
plt.show()

# =========================
# 3. MODEL RANKING
# =========================
sorted_models = sorted(results.items(), key=lambda x: x[1]["accuracy"], reverse=True)

names = [m[0] for m in sorted_models]
scores = [m[1]["accuracy"] for m in sorted_models]

plt.figure()
plt.bar(names, scores)

plt.xticks(rotation=20)
plt.ylabel("Accuracy")
plt.title("Model Ranking Based on Accuracy")

plt.tight_layout()
plt.savefig("images/model_ranking.png")
plt.show()

plt.figure()
df['sentiment'].value_counts().plot(kind='bar')

plt.title("Dataset Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")

plt.tight_layout()
plt.savefig("images/distribution.png")
plt.show()
```
